```python
import pandas as pd
import os
```

```python
#Import dữ liệu
path="./raw-data"
files=[file for file in os.listdir(path) if not file.startswith(".")]
all_books=pd.DataFrame()

for file in files:
    current_data=pd.read_csv(path+"/"+file)
    all_books=pd.concat([all_books,current_data])
    
all_books.to_csv("all_tiki_books.csv",index=False,encoding='utf-8-sig')
```

```python
all_data = pd.read_csv("all_tiki_books.csv")
#Extract các cột data cần
new_tikibook = all_data.filter(['book-rank','book-link-href','Name','Author','Category','Price','Rating-number','Sold-number'], axis=1)
new_tikibook.to_csv('new-tikibook.csv',index=False,encoding='utf-8-sig')
new_tikibook.head()
```

```python
new_tikibook=pd.read_csv("./new-tikibook.csv")
new_tikibook.head()
```

### Extract ID sản phẩm (PID) từ cột 'book-link-href'

```python
def get_PID_link(productlink):
    return productlink.split(".",2)[1]

new_tikibook['subID'] = new_tikibook['book-link-href'].apply(lambda x: f"{get_PID_link(x)}")
new_tikibook['subID'].head(50)
```

```python
#Lấy số ID mỗi sản phẩm
def get_PID(productid):
    return productid.split("-")[-1]
new_tikibook['PID'] = new_tikibook['subID'].apply(lambda x: f"{get_PID(x)}")
new_tikibook['PID'] = new_tikibook['PID'].map(lambda x: x.lstrip('p'))
new_tikibook['PID'].head(50)
```

```python
#Vì PID của một số sản phẩm tách chuỗi bị lỗi và có giá trị là "tiki". Cần tạo dataframe mới chứa các sản phẩm PID="tiki" để edit lại
no_PID = new_tikibook[new_tikibook['PID']=="tiki"]
no_PID.head(50)
```

```python
#Bỏ từ "tiki" khỏi ID sản phẩm
new_tikibook.drop(new_tikibook[new_tikibook['PID'] =="tiki"].index, inplace = True)
new_tikibook.head(50)
```

```python
#Tạo 1 dataframe no_PID để trích xuất các link cùng định dạng và có cách tách ký tự và lấy PID giống nhau
def get_PID_link(productlink):
    return productlink.split(".")[-2]

no_PID['subID'] = no_PID['book-link-href'].apply(lambda x: f"{get_PID_link(x)}")
no_PID.head()
```

```python
#Tách PID với định dạng pXXXXX từ value trong cột 'subID'
def get_PID(productid):
    return productid.split("-")[-1]
no_PID['PID'] = no_PID['subID'].apply(lambda x: f"{get_PID(x)}")

#Tách ký tự 'p' trong PID:
no_PID['PID'] = no_PID['PID'].map(lambda x: x.lstrip('p'))
no_PID['PID'].head(50)
```

```python
new_tikibook_with_ID = pd.concat([new_tikibook, no_PID], ignore_index=True, sort=False)
new_tikibook_with_ID.to_csv("new_tikibook_with_ID.csv",index=False,encoding='utf-8-sig')
```

```python
new_tikibook_with_ID=new_tikibook_with_ID.drop('subID', 1)
```

```python
new_tikibook_with_ID['Rating-number'] = new_tikibook_with_ID['Rating-number'].fillna(0)
new_tikibook_with_ID['Rating-number'] = new_tikibook_with_ID['Rating-number'].str.replace(r'\D', '')
new_tikibook_with_ID.dtypes
```

```python
# Xóa các giá trị null trong cột 'Sold-number'
new_tikibook_with_ID['Sold-number'] = new_tikibook_with_ID['Sold-number'].fillna(0)

# Xóa ký tự str trong cột 'Sold-number'
new_tikibook_with_ID['Sold-number'] = new_tikibook_with_ID['Sold-number'].str.replace(r'\D', '')
new_tikibook_with_ID.head()
```

```python
#Định dạng Cột 'Sold-number' từ string thành float
new_tikibook_with_ID['Sold-number'] = new_tikibook_with_ID['Sold-number'].astype(float)
```

```python
#Định dạng Cột 'Rating-number' từ string thành float
new_tikibook_with_ID['Rating-number'] = new_tikibook_with_ID['Rating-number'].astype(float)
```

```python
new_tikibook_with_ID.dtypes
```

```python
new_tikibook_with_ID['Price'] = new_tikibook_with_ID['Price'].str.replace(r'\D', '').astype(float)
```

```python
new_tikibook_with_ID = new_tikibook_with_ID[['PID','book-rank','book-link-href','Name','Author','Category','Price','Rating-number','Sold-number']]
new_tikibook_with_ID.to_csv("new_tikibook_with_ID.csv",index="False",encoding='utf-8-sig')
```
