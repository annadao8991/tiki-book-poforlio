```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Thể loại sách nào phổ biến/bán chạy/được nhập nhiều nhất?
### What category is most popular/best-seller/newest?

### 1a. Sách thể loại nào phổ biến nhất ? - Most popular book catergory?

```python
popular_books=all_books.loc[all_books['book-rank'] == "Phổ biến"]
popular_books
```

```python
popular_books_cat=popular_books.groupby('Category').count()['PID']
popular_books_cat
```

```python
cats = [cat for cat, df in popular_books.groupby(['Category'])]

plt.bar(cats,popular_books.groupby(['Category']).count()['PID'])
plt.ylabel('Number of Popular books per Category ($)')
plt.xlabel('Category')
plt.xticks(cats, rotation='vertical', size=8)
plt.show()
```

#### 1a - Trả lời:
#### - Top 5 thể loại sách phổ biến nhất bao gồm:
#### Văn học (66) > Kỹ năng sống (46) > Kinh tế (35) > Ngoại ngữ (15) > Thiếu nhi (11)
#### Trong đó, thể loại Văn học là phổ biến nhất và cách biệt khá xa các loại sách khác.

```python
best_seller=all_books.loc[all_books['book-rank'] == "Bán chạy"]
best_seller
```

```python
cats = [cat for cat, df in best_seller.groupby(['Category'])]

plt.bar(cats,best_seller.groupby(['Category']).count()['PID'])
plt.ylabel('Number of Best-seller books per Category')
plt.xlabel('Category')
plt.xticks(cats, rotation='vertical', size=8)
plt.show()
```

```python
newest_book=all_books.loc[all_books['book-rank'] == "Hàng mới"]
cats = [cat for cat, df in newest_book.groupby(['Category'])]

plt.bar(cats,newest_book.groupby(['Category']).count()['PID'])
plt.ylabel('Number of Newest books per Category')
plt.xlabel('Category')
plt.xticks(cats, rotation='vertical', size=8)
plt.show()
```

Câu 1 - Trả lời:
*Top 5 thể loại sách Phổ biến nhất bao gồm:

Văn học > Kỹ năng sống > Kinh tế > Ngoại ngữ > Thiếu nhi

*Top 5 thể loại sách Bán chạy nhất bao gồm:

Kinh tế > Kỹ năng sống = Văn học > Thiếu nhi > Ngoại ngữ

*Top 5 thể loại sách Nhập nhiều nhất bao gồm:

Thiếu nhi > Tham khảo > Văn học > Kinh tế > Ngoại ngữ.

### 1. Sách phổ biến

#### Doanh thu sách phổ biến theo từng thể loại

```python
popular_books
```

```python
popular_books['Product Sales']=popular_books['Price']*popular_books['Sold-number']
popular_books
```

```python
popular_books_sales=popular_books.groupby(['Category']).sum()['Product Sales']
popular_books_sales
```

```python
from matplotlib.ticker import ScalarFormatter
```

```python
cats = [cat for cat, df in popular_books.groupby(['Category'])]

plt.bar(cats,popular_books_sales)
plt.ylabel('Total Sales of popular books per Category (VND)')
plt.xlabel('Category')
plt.xticks(cats, rotation='vertical', size=8)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
